```python
import pandas as pd
import re


df = pd.read_csv('Эвристика_выручка_недели_1.csv', sep=';')[4:-2]
df.columns = df.head(1).values[0]
df = df[2:]
df.fillna('0', inplace=True)
df.rename(columns={'Названия строк':'date', 'Общий итог':'sum'}, inplace=True)
df.reset_index(drop=True, inplace=True)
    
def to_num(string):
    try:
        position = re.search(r',\d\d', string).start()
        string = string[:position].replace(' ', '')
        return int(string)
    except: return int(string)

for num in df:
    df[num] = df[num].apply(to_num)
df
```

```python
def week_3(series):
    week_avg = []

    for i in df.index:
        if i == 0:
            week_avg.append(series[i])
        elif i == 1: week_avg.append(int((series[i] + series[i-1])/2)) 
        else: week_avg.append(int((series[i] + series[i-1] + series[i-2])/3)) 
    return pd.Series(week_avg)

dict_3 = {}
for column in df.columns:
    dict_3[column] = week_3(df[column])
    
pd.DataFrame(dict_3)
```

```python
def week_tend_columns(series):
    week_tend = []
    

    for i in df.index:
        if i == 0:
            week_tend.append(1)
        else: week_tend.append(series[i]/series[i-1])
    return pd.Series(week_tend)

week_tend_columns(df['sum'])
```

# Влияние предыдущего года

Усредненное по трем неделям предыдущего года аналогичного периода. Cовместить с тенденцией к изменению

```python
def year_season(column):
    column = week_tend_columns(column)
    year_season = []
    for i in df.index:
        try: year_season.append(
            (column[i-53] + column[i-52] + column[i-51])/3)

        except: year_season.append(None)
    return pd.Series(year_season)
```

Для того, чтобы минимизировать влияние

Формула получается простая: среднее по трем крайним неделям * на сезонный коэффициент.
- Где сезонный коэффициент является средним изменением поставок за крайний год

Минусом является то, что поставщик может вести себя как сезонно, так и хаотично. А для того, чтобы сделать дополнительный коэффициент степени сезонности поставщика по его истории нужнен не год наблюдений.

```python
cont = df['sum']

result_formula = ((year_season(cont) * week_3(cont))).dropna().apply(lambda x: int(x))

result_nowday = week_3(cont)[53:]

print('Результат моей формулы:  ',(result_formula-cont[53:]).abs().sum())
print('Результат старой формулы:',(result_nowday-cont[53:]).abs().sum())
```
